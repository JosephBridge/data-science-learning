## Apply `lambda` function

To a column:

```python
df = pd.read_csv('employees.csv')

get_last_name = lambda name: name.split(' ')[-1]

df['last_name'] = df.name.apply(get_last_name)
df.head()
```

To a row:

```python
total_earned = lambda row: row['hours_worked'] * \
           row['hourly_wage'] if row['hours_worked'] <= 40 else (40 * row['hourly_wage'])\
            + (row['hours_worked'] - 40) * \
            (row['hourly_wage'] * 1.50)

df['total_earned'] = df.apply(total_earned, axis=1)
df.head()
```
