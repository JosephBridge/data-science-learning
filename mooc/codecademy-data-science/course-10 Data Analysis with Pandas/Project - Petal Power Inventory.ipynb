```python
import pandas as pd

# read the data
inventory = pd.read_csv("inventory.csv")

# check the 1st 10 rows
inventory.head(10)
```

```python
# 1st 10 rows
staten_island = inventory.iloc[:10]
staten_island
```

```python
product_request = staten_island['product_description']
product_request
```

```python
seed_request = staten_island[(staten_island.location=='Brooklyn') & (staten_island.product_type=='seeds')]
seed_request
```

```python
# add a new column
inventory['in_stock'] = inventory.quantity.apply(lambda num: True if num>0 else False)
inventory.head()
```

```python
# other way
inventory['in_strock_2'] = inventory.apply(lambda row: True if row.quantity>0 else False, axis=1)
inventory.head()
```

```python
inventory['total_value'] = inventory.price * inventory.quantity
inventory.head()
```

```python
# The following lambda function combines product_type and product_description into a single string:
combine_lambda = lambda row: \
    '{} - {}'.format(row.product_type,
                     row.product_description)
```

```python
inventory['full_description'] = inventory.apply(combine_lambda, axis=1)
inventory.head()
```
